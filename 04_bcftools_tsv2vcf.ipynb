# Convert 23andme to VCF

## convert lifted_build36 file

```python
ref_file = '/home/tun01458/lib/ref_genome/Homo_sapiens_assembly19.fasta'
lifted_build36_dir = 'user_data/lifted_build36_to_build37/'
individual_vcf_dir = 'user_data/individual_genotype_vcf/'

for i, user_genotype in enumerate(os.listdir(lifted_build36_dir)):
    print('{}/{}'.format(i + 1, len(os.listdir(lifted_build36_dir))))
    print(user_genotype)

    user_id = user_genotype.split('_')[0][4:]
    args = [
        'bcftools', 'convert', '-c', 'ID,CHROM,POS,AA', '-f', ref_file,
        '--tsv2vcf',
        os.path.join(lifted_build36_dir,
                     user_genotype), '-s', user_id, '-Oz', '-o',
        os.path.join(individual_vcf_dir, user_genotype + '.vcf.gz')
    ]
    command_line = ' '.join(args)
    print(command_line)
    subprocess.call(command_line, shell=True)
```

# sort SNP position

```python
# sort file
individual_vcf_dir = 'user_data/individual_genotype_vcf/'
sorted_individual_vcf_dir = 'user_data/sorted_individual_genotype_vcf/'

# bcftools sort individual_genotype_vcf/user1005_file541_yearofbirth_1981_sex_XY.23andme.txt.vcf.gz -Oz -o sorted_individual_genotype_vcf/user1005_file541_yearofbirth_1981_sex_XY.23andme.txt.vcf.gz

# generate new .csi files
individual_vcf_list = os.listdir(individual_vcf_dir)
for i, user_genotype in enumerate(individual_vcf_list):
    print('{}/{}'.format(i + 1, len(individual_vcf_list)))
    print(user_genotype)

    args = [
        'bcftools', 'sort',
        os.path.join(individual_vcf_dir, user_genotype), '-Oz', '-o',
        os.path.join(sorted_individual_vcf_dir, user_genotype)
    ]
    command_line = ' '.join(args)
    print(command_line)
    subprocess.call(command_line, shell=True)
```

# merge into one VCF file

```python
# generate index file
sorted_individual_vcf_dir = 'user_data/sorted_individual_genotype_vcf/'

sorted_individual_vcf_list = glob.glob(sorted_individual_vcf_dir + '*.vcf.gz')
for i, user_genotype in enumerate(sorted_individual_vcf_list):
    print('{}/{}'.format(i + 1, len(sorted_individual_vcf_list)))
    print(user_genotype)

    args = ['bcftools', 'index', '-f', user_genotype]
    command_line = ' '.join(args)
    print(command_line)
    subprocess.call(command_line, shell=True)
```

```python
# merge genotypes
sorted_individual_vcf_dir = 'user_data/sorted_individual_genotype_vcf/'
args = [
    'bcftools', 'merge', '--force-samples',
    sorted_individual_vcf_dir + '*.vcf.gz', '-Oz', '-o',
    'user_data/lactose_intolerant.vcf.gz'
]
command_line = ' '.join(args)
print(command_line)
subprocess.call(command_line, shell=True)
```
