# Introduction

User genotype data was built based on two references build36 (hg18) and build37 (hg19). First we need to separate different type files. Then lift over the files based on build36 to build37.

This notebook is to separate different type files.

```python
import os
import json
import shlex, subprocess
import pandas as pd
```

# Separate different building data

Separate different build36 and build37 data.

```python
# check build reference and transfer file format to bed
count_error = 0
count_build36 = 0
count_build37 = 0

user_genotype_dir = 'user_data/original_lactose_genotype/'
for user_genotype_filename in os.listdir(user_genotype_dir):
    print(user_genotype_filename, end=' ')
    try:
        # check which build references
        with open(os.path.join(user_genotype_dir, user_genotype_filename),
                  'r') as fh_in:
            for _ in range(11):
                next(fh_in)
            if 'human assembly build 36' in fh_in.readline():
                new_file_dir = 'build36_genotype'
                count_build36 += 1
            else:
                new_file_dir = 'build37_genotype'
                count_build37 += 1
        print(new_file_dir)

        args = [
            'cp',
            os.path.join('user_data/original_lactose_genotype/',
                         user_genotype_filename),
            os.path.join('user_data/', new_file_dir)
        ]
        command_line = ' '.join(args)
        # print(command_line)
        subprocess.call(command_line, shell=True)
    except:
        count_error += 1
        print('Error')
        pass

count_build36, count_build37, count_error
```
