# Introduction

The lactose_intolerant.json file contains all users that annotations their phenotype of lactose tolerant. This program is to extract user id and phenotype.

```python
import os
import json
import shlex, subprocess
import pandas as pd
```

# Parsing json file

Parsing json file to get users' id and phenotype

```python
json_file = 'user_data/lactose_intolerant.json'

users_id = []
phenotype = []
with open(json_file, 'r') as read_json:
    data = json.load(read_json)
#     print(data)
    # extract user_id from users list
    for user in data['users']:
#         print(user)
        print(user['user_id'], user['variation'])

        users_id.append(str(user['user_id']))
        phenotype.append(user['variation'])
```

```python
len(users_id), len(phenotype)
```

```python
all_user_filename = os.listdir('/home/tun01458/data/openSNP/opensnp_datadump/data/')

# only using 23andme data
user_23andme_filename = [
    user_filename for user_filename in all_user_filename
    if (user_filename.split('.')[-2] == '23andme')
    & (user_filename.split('_')[0][4:] in users_id)
]

print(len(user_23andme_filename))
user_23andme_filename
```

# Check file size

```python
# filesize belongs in [10, 40] MB
file_dir = '/home/tun01458/data/openSNP/opensnp_datadump/data/'
eligible_files = [
    file for file in user_23andme_filename
    if (os.path.getsize(os.path.join(file_dir, file)) < 40 * 1024 * 1024)
    & (os.path.getsize(os.path.join(file_dir, file)) > 10 * 1024 * 1024)
]

print(len(eligible_files))
eligible_files
```

# copy valid file

```python
# example: cp /path/{a,b,c} /path/to/
ori_dir = '/home/tun01458/data/openSNP/opensnp_datadump/data/'
dst_dir = 'user_data/original_lactose_genotype/'

args = ['cp', ori_dir + '{' + ','.join(eligible_files) + '}', dst_dir]
command_line = ' '.join(args)
print(command_line)
subprocess.call(command_line, shell=True)
```
